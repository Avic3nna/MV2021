```python
# take picture from video stream on pressing escape key
# pictures are taken with 2 top horizontal light strips and the table light towards the administration door (with blinders closed)
import cv2

cv2.namedWindow("tripod cam")
vc = cv2.VideoCapture(1)

if vc.isOpened(): # try to get the first frame
    rval, frame = vc.read()
else:
    rval = False

i=1
while rval:
    cv2.imshow("tripod cam", frame)
    rval, frame = vc.read()
    key = cv2.waitKey(20)
    if key == 27: # exit on ESC
        break
    if key == 32: #exit on spacebar
        path = 'Images/Templates/'
        ext = '.png'

        img_path = path+str(i)+ext
        cv2.imwrite(img_path, frame)
        i += 1


vc.release()
cv2.destroyWindow("tripod cam")
```

```python
# Live detection

template = cv2.imread(r"G:\\My Drive\\1. EIT Digital master\\Estland\\Semester 1\\Machine vision\\MV2021\\Homework 3\\Python OpenCV2\\Images\\Templates\\key_roi.png")
#image_RGB = cv2.imread(r"G:\\My Drive\\1. EIT Digital master\\Estland\\Semester 1\\Machine vision\\MV2021\\Homework 3\\Python OpenCV2\\Images\\key_pic_3.png")

cv2.namedWindow("tripod cam")
vc = cv2.VideoCapture(1)

if vc.isOpened(): # try to get the first frame
    rval, img = vc.read()
else:
    rval = False

cv2.destroyWindow("tripod cam")

while True:
    rval, img = vc.read()
    key = cv2.waitKey(20)
    
    result = cv2.matchTemplate(img, template, cv2.TM_CCOEFF_NORMED)

# show the output image
    probThreshold = 0.15
    h, w = template.shape[:2]
    # fake out max_val for first run through loop
    new_im = img
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    if max_val > probThreshold:
        result[max_loc[1]-h//2:max_loc[1]+h//2+1, max_loc[0]-w//2:max_loc[0]+w//2+1] = 0
        new_im = cv2.rectangle(img,(max_loc[0],max_loc[1]), (max_loc[0]+w+1, max_loc[1]+h+1), (0,255,0) )   

    #cv2.imwrite('output.png', new_im)

    cv2.imshow("Output", new_im)
    cv2.waitKey(20)
        

    if key == 27: # exit on ESC
        break

vc.release()
cv2.destroyWindow("Output")
```
